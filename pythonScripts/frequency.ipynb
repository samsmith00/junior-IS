```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4*np.pi + 3, 100)

y = np.cos(x)

plt.plot(x, y)
plt.title("Cosine Wave Over a Period of 4π")
plt.xlabel("Samples")
plt.ylabel("Magnitude")
plt.show()

plt.magnitude_spectrum(y)
plt.title("Frequency Spectrum Plot")
plt.xlabel("Frequency (arb. units)")
plt.ylabel("Magnitude")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = 7     
fs = 1000    
N = 1000  

t = np.arange(N) / fs  

y = np.cos(2 * np.pi * f * t)

plt.plot(t, y)
plt.title("Cosine Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

X = np.fft.rfft(y)
freqs = np.fft.rfftfreq(N, 1/fs)

X_mag = np.abs(X) / N   
X_db = 20 * np.log10(X_mag + 1e-12)

plt.plot(freqs, X_db)
plt.title("Spectrum Leakage")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude (dB)")
plt.ylim(-120, 5)   
plt.xlim(0, fs/2)   
plt.show()


```
