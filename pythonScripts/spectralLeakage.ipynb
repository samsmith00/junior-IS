```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# pure cos wav, no leakage
x = np.arange(0, 50*np.pi, 0.01)
y = 50 * np.cos(x)

plt.plot(x, y)
```

```python
freqs, mags, _ = plt.magnitude_spectrum(y, window=None)
plt.xlim(0, 0.1)  # adjust this upper limit to taste
plt.tight_layout()
plt.show()
```

```python
# composition wave
x2 = np.linspace(0, 125*np.pi, 2048)
y2 = np.cos(0.2*x2) + np.cos(0.35*x2)

plt.title("Rectangle Window")
plt.ylabel("Amplitude")
plt.xlabel("Samples")
plt.plot(x2, y2)
```

```python
# composition wave
x2 = np.linspace(0, 125*np.pi, 2048)
y2 = np.cos(0.2*x2) + np.cos(0.35*x2)

y2 *= np.hanning(len(x2))

plt.title("Hanning Window")
plt.ylabel("Amplitude")
plt.xlabel("Samples")
plt.plot(x2, y2)
```

```python


N_fft = len(x2) * 8

window_signal = y2 * np.hanning(len(x2))

fft_vals_no_win = np.fft.fft(y2, n=N_fft)
fft_vals_win = np.fft.fft(window_signal, n=N_fft)


no_window_db = 20 * np.log10(np.abs(fft_vals_no_win[:N_fft//2]) + 1e-10)
window_db = 20 * np.log10(np.abs(fft_vals_win)[:N_fft//2] + 1e-10)


plt.title("Rectangle Windowed FFT")
plt.ylabel("Magnitude (dB)")
plt.xlabel("frequency bins")
plt.plot(no_window_db)
plt.xlim(0, 2000)
plt.show()

plt.title("Hanning Windowed FFT")
plt.ylabel("Magnitude (dB)")
plt.xlabel("frequency bins")
plt.plot(window_db)
plt.xlim(0, 2000)
plt.show()

```
