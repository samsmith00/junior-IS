```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# pure cos wav, no leakage
x = np.arange(0, 4*np.pi, 0.01)
y = 50 * np.cos(x)

plt.plot(x, y)
```

```python
freqs, mags, _ = plt.magnitude_spectrum(y, window=None)
plt.xlim(0, 0.1)  # adjust this upper limit to taste
plt.tight_layout()
plt.show()
```

```python
# composition wave
x2 = np.arange(0, 5*np.pi, 0.1)
y2 = 50 * np.cos(x2)


plt.plot(x2, y2)
```

```python
freqs, mags, _ = plt.magnitude_spectrum(y2, window=None)
plt.xlim(0, 0.1)  # adjust this upper limit to taste
plt.tight_layout()
plt.show()
```
